```python
import pandas as pd

df = pd.read_csv('C:/Users/SABIO/Documents/GitHub/IE-University/CSV_FILES/Student_Performance_Data.csv')

df.head(3)
```

Find the average Test_Score for each class. Sort the results in descending order of Average_Score. Display the result as a DataFrame with Class and Average_Score rounded to 1 decimal place.

```python
df.groupby('Class', as_index = False)['Test_Score'].mean().round(1).sort_values(by = 'Test_Score', ascending = False)
```

Identify the top 5 students with the highest Attendance_Rate. If there are ties, sort by Test_Score in descending order. Display the following columns: Student_ID, Class, Attendance_Rate, and Test_Score.

```python
df.sort_values(by = ['Attendance_Rate', 'Test_Score'], ascending = [False, False])[['Student_ID', 'Class', 'Attendance_Rate', 'Test_Score']].head(5)
```

For each class, count how many students have submitted more than 10 assignments. Display the Class and the count of students as Students_Submitted_More_Than_10.

```python
result = df[df['Assignments_Submitted'] > 10].groupby('Class', as_index = False).size()

result.columns = ['Class', 'Students_Submitted_More_Than_10']

result
```

Find the median Test_Score grouped by Age and Class. Sort the results first by Age (ascending) and then Class alphabetically. Display Age, Class, and Median_Score.

```python
result = df.groupby(['Age', 'Class'], as_index = False)['Test_Score'].median().sort_values(by = ['Age', 'Class'], ascending = [True, True])

result.columns = ['Age', 'Class', 'Median_Score']

result
```

Find the student with the lowest Attendance_Rate in each class. Display Class, Student_ID, Attendance_Rate, and Enrollment_Date.

```python
df.loc[df.groupby('Class')['Attendance_Rate'].idxmin(), ['Class', 'Student_ID', 'Attendance_Rate', 'Enrollment_Date']]
```
