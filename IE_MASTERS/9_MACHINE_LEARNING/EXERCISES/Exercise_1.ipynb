### Exercise 1 - Data manipulation

```python
import pandas as pd
import numpy as np

# Load the datasets
custtravel_path = 'Custtravel.dat'
holtravel_path = 'Holtravel.dat'

# Read data
custtravel = pd.read_csv(custtravel_path, sep=';')
holtravel = pd.read_csv(holtravel_path, sep=';')

custtravel.info()
print("")
holtravel.info()
```

```python
# 1. Combine datasets
custtravel.rename(columns={"HOLCODE": "holidaycode"}, inplace=True)
custandhol = pd.merge(custtravel, holtravel, on="holidaycode", how="inner")
custandhol.info()
```

```python
pd.set_option('display.max_columns', None)
custandhol
```

```python
# 2. Correct variable features
custandhol.rename(columns={"DOB": "Date of Birth"}, inplace=True)
custandhol["Date of Birth"] = pd.to_datetime(custandhol["Date of Birth"], format="%d/%m/%Y", errors='coerce')
custandhol["TRAVDATE"] = pd.to_datetime(custandhol["TRAVDATE"], format="%d/%m/%Y", errors='coerce')
custandhol["GENDER"] = custandhol["GENDER"].map({"0": "Female", "1": "Male"}).fillna("Unknown")
custandhol["ACCOMMODATION"].fillna("NA")
custandhol.info()
```

```python
# 3. Create new variables
# a1) Remove outliers from HOLCOST
custandhol["HOLCOST"] = custandhol["HOLCOST"].str.replace("$null$", "")
custandhol["HOLCOST"] = pd.to_numeric(custandhol["HOLCOST"], errors="coerce")
q1 = custandhol["HOLCOST"].quantile(0.25)
q3 = custandhol["HOLCOST"].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
custandhol = custandhol[(custandhol["HOLCOST"] >= lower_bound) & (custandhol["HOLCOST"] <= upper_bound)]
```

```python
custandhol.shape
```

```python
# a2) Standardize
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
custandhol.loc[:,"Z_HOLCOST"] = scaler.fit_transform(custandhol[["HOLCOST"]])
```

```python
custandhol
```

```python
# b) Extract month from Date of Birth, calculate customer AGE and remove records with a value < 18 years
custandhol.loc[:,"Month"] = custandhol["Date of Birth"].dt.month
current_year = pd.Timestamp.now().year
custandhol.loc[:,"AGE"] = current_year - custandhol["Date of Birth"].dt.year
custandhol = custandhol[custandhol["AGE"] >= 18]
```

```python
# c) Create 0-1 flag variables for COUNTRY
country_dummies = pd.get_dummies(custandhol["COUNTRY"], prefix="COUNTRY", dtype=int)
custandhol = pd.concat([custandhol, country_dummies], axis=1)
```

```python
custandhol
```

```python
# d) Create WEEKEND_SPAIN
custandhol["WEEKDAY"] = custandhol["TRAVDATE"].dt.weekday+1
custandhol["WEEKEND_SPAIN"] = ((custandhol["WEEKDAY"] >= 5) & (custandhol["WEEKDAY"] < 7) & (custandhol["COUNTRY"] == "SPAIN"))
```

```python
# e) Transform HOLCOST
custandhol["LN_HOLCOST"] = np.log(custandhol["HOLCOST"] + 1)  # Avoid log(0)
custandhol["SQRT_HOLCOST"] = np.sqrt(custandhol["HOLCOST"])
custandhol["INV_HOLCOST"] = 1 / (custandhol["HOLCOST"] + 1)  # Avoid division by 0
```

```python
# 4. Extract surname from NAME
custandhol["SURNAME"] = custandhol["NAME"].str.split().str[-1]
```

```python
custandhol
```

```python
# 5. Create RAND variable and mark cases
custandhol["RAND"] = np.random.rand(len(custandhol))
custandhol["RAND_MARK"] = custandhol["RAND"] > 0.2
```

```python
# 7. Which variable holds more information, either Nights or Distbeach?
varlist=["NIGHTS", "DISTBEACH"]

for var in varlist:
    mean = np.mean(custandhol[var])
    std_dev = np.std(custandhol[var])
    varcoef=std_dev/mean
    print ("Variation coefficient of", var, "is", varcoef)
```

```python
# 8a. Discrete version of HOLCOST variable 
var="HOLCOST"
n_bins = 4
bin_edges = np.linspace(min(custandhol[var]), max(custandhol[var]), n_bins+1)
discrete_data = pd.cut(custandhol[var], bins=bin_edges, labels=False)

discrete_data.name="HOLCOST_discrete1"
custandhol = pd.concat([custandhol, discrete_data], axis=1)

print("Bin size:", bin_edges[1] - bin_edges[0])
print("Bin Edges:", bin_edges)
```

```python
# 8b. How many records are equal or greater than 2000?
bins = [0, 1000, 2000, 3000, 4000, 5000] 

discrete_data = np.digitize(custandhol[var], bins)

bin_counts = np.bincount(discrete_data - 1, minlength=len(bins))

import matplotlib.pyplot as plt

bin_labels = [f"{bins[i]}-{bins[i+1]}" for i in range(len(bins) - 1)]
plt.bar(bin_labels, bin_counts[:-1], edgecolor="black", alpha=0.7)
for i, count in enumerate(bin_counts[:-1]):
    plt.text(i, count + 0.1, str(count), ha='center', va='bottom', fontsize=10)
plt.xlabel("Bins")
plt.ylabel("Count")
plt.title("Element Count in Each Bin")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
# 9. Combine variables GENDER and COUNTRY. Do you see any difference between genders?

var1 = custandhol["GENDER"]
var2 = custandhol["COUNTRY"]

var1 = custandhol["GENDER"][(custandhol["GENDER"] != "Unknown")] # to filter out Unknown

from itertools import product
combinations = list(product(var1, var2))

df = pd.DataFrame(combinations, columns=['Var1', 'Var2'])

frequencies = np.random.randint(1, 10, size=len(df))  # Generar frecuencias aleatorias
df['Frequency'] = frequencies

frequency_table = df.pivot_table(index='Var1', columns='Var2', values='Frequency', aggfunc='sum')
percentage_table = frequency_table.div(frequency_table.sum(axis=0), axis=1) * 100

print("Frequencies ")
print(frequency_table)
print("")
print("Column Percentages")
print(percentage_table)
```

```python
#11 Bin the variable HOLCOST using the ACCOMMODATION var with values either APARTMENT OR NOT

custandhol['ACCOMM2'] = custandhol['ACCOMMODATION'].apply(lambda x: 0 if x in ['HOTEL', 'CABIN'] else 1)

variable = "HOLCOST"
x = custandhol[variable].values
y = custandhol["ACCOMM2"]

from optbinning import OptimalBinning

optb = OptimalBinning(name=variable, dtype="numerical", solver="cp")
optb.fit(x, y)
optb.status
optb.binning_table.build()
optb.binning_table.plot(metric="woe") # Weight of evidence: tells the predictive power of an independent variable in relation to the dependent one
optb.binning_table.analysis()
```
