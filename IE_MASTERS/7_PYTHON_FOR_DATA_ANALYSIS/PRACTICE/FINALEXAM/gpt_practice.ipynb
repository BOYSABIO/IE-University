Find the average salary of employees in each department. The result should be a DataFrame with two columns: Department and Average_Salary. Round to 2 decimal places.

```python
employee.groupby('Department')['Salary'].mean().round(2)
```

Get the top 5 employees with the highest performance scores. If there are ties, sort by Salary in descending order. Display Employee_ID, Department, Performance_Score, and Salary.

```python
employee.groupby('Employee_ID')[['Department', 'Performance_Score', 'Salary']].sum().sort_values(by = 'Performance_Score', ascending = False).head(10)
```

```python
result = employee.sort_values(by = ['Performance_Score', 'Salary'], ascending = [False, False])
result = result[['Employee_ID', 'Department', 'Performance_Score', 'Salary']].head(5)
result
```

How many employees in each department have more than 10 years of experience? Show the department name and count of employees.

```python
employee[employee['Experience_Years'] > 10].groupby('Department', as_index = False)['Employee_ID'].size()
```

What is the median age of employees grouped by Performance_Score? The result should display Performance_Score and Median_Age.

```python
employee.groupby('Performance_Score')['Age'].median()
```

Find the oldest employee in each department. Display Department, Employee_ID, Age, and Hire_Date.

```python
employee.groupby(['Department', 'Employee_ID'])[['Age', 'Hire_Date']].agg({'Age':'max', 'Hire_Date':'sum'}).sort_values(by = 'Age', ascending = False)
```

```python
result = employee.loc[employee.groupby('Department')['Age'].idxmax(), ['Department', 'Employee_ID', 'Age', 'Hire_Date']]
result
```
