```python
#With the data available, who is the fastest driver in Formula 1? (0.5 point)
#Based on average fastest lap speed
average_race_speed = df.groupby('driver')['fastest_lap_speed'].mean()

sorted_average_race_speed = average_race_speed.sort_values(ascending=False)

sorted_average_race_speed.head()
```

```python
#Based on fastest lap time

import pandas as pd
import numpy as np

def time_to_seconds(time_value):
    if pd.isnull(time_value):
        return np.nan
    if isinstance(time_value, (int, float)):
        return time_value
    time_str = str(time_value)

    if ':' in time_str:
        minutes, seconds = time_str.split(':')
        total_seconds = int(minutes) * 60 + float(seconds)
        return total_seconds
    else:
        # Handle cases without a colon (just seconds)
        return float(time_str)


df['fastest_lap_seconds'] = df['fastest_lap_time'].apply(time_to_seconds)
sorted_df = df[df['fastest_lap_seconds'].notna()]

print('All drivers and thier fastest times grouped by circuits:')
print(sorted_df.groupby(['circuit_name', 'driver'])['fastest_lap_seconds'].min().sort_values(ascending = True))
print()

sorted_df = df[df['fastest_lap_seconds'].notna()]

driver_lap_times = sorted_df.groupby(['circuit_name', 'driver'])['fastest_lap_seconds'].min().reset_index()

driver_lap_times['min_time_per_circuit'] = driver_lap_times.groupby('circuit_name')['fastest_lap_seconds'].transform('min')

fastest_drivers = driver_lap_times[driver_lap_times['fastest_lap_seconds'] == driver_lap_times['min_time_per_circuit']]

fastest_drivers = fastest_drivers.drop(columns='min_time_per_circuit')

print('The fastest driver and their fastest time grouped by circuit:')
sorted_fastest_drivers = fastest_drivers[['circuit_name', 'driver', 'fastest_lap_seconds']].sort_values(by = 'fastest_lap_seconds', ascending = True)
print(sorted_fastest_drivers.head(3))
print()

print('The number of times a drver appears as the fastest across all circuits:')
most_appearing_drivers = sorted_fastest_drivers['driver'].value_counts()
print(most_appearing_drivers.head(3))
print()

print(f'{most_appearing_drivers.idxmax()} appears the most as the fastest driver across {most_appearing_drivers.max()} different circuits.')
```

```python
# Which is the driver with the most podiums without a win (position 2 or 3)? (0.5 point)
podium_finishes = df[df['final_position'].isin([2, 3])]

drivers_no_wins = df.groupby('driver')['final_position'].apply(lambda x: 1 not in x.values)

podium_no_wins = podium_finishes[podium_finishes['driver'].isin(drivers_no_wins[drivers_no_wins].index)]

podium_counts = podium_no_wins['driver'].value_counts()

print(podium_counts.head(3))

print(f'{podium_counts.idxmax()} has the most podiums ({podium_counts.max()}) without a win.')
```

```python
# Number of races per driver

df_cleaned = df[['driver']].copy()
df_cleaned = df_cleaned.dropna()

driver_row_counts_corrected = df_cleaned['driver'].value_counts().reset_index()
driver_row_counts_corrected.columns = ['driver', 'row_count']
driver_row_counts_corrected_sorted = driver_row_counts_corrected.sort_values(by='row_count', ascending=False).reset_index(drop=True)

print('Number of races per driver:')
print(driver_row_counts_corrected_sorted.head())
```

```python
# Number of times driver beat their teammate

df_filtered = df[['driver', 'constructor_name', 'race_name', 'final_position', 'year']].copy()
df_filtered['final_position'] = pd.to_numeric(df_filtered['final_position'], errors='coerce').fillna(24) # Replace NaN in 'final_position' with 24 (for non-finishers)

df_filtered['unique_race'] = df_filtered['race_name'].astype(str) + "_" + df_filtered['year'].astype(str)

def calculate_better_finishes(group):
    group = group.sort_values('final_position').reset_index(drop=True)
    group['better_than_teammate'] = group['final_position'] < group['final_position'].shift(-1)
    return group

df_comparison = df_filtered.groupby(['unique_race', 'constructor_name']).apply(calculate_better_finishes)

driver_better_counts = df_comparison.groupby('driver')['better_than_teammate'].sum().reset_index()

driver_better_counts.columns = ['driver', 'better_than_teammate_count']
driver_better_counts = driver_better_counts.sort_values(by='better_than_teammate_count', ascending=False).reset_index(drop=True)

print(driver_better_counts.head(10))
```

```python
driver_combined = driver_row_counts_corrected_sorted.merge(driver_better_counts, on='driver', how='inner')


driver_combined['percentage_above_teammate'] = (driver_combined['better_than_teammate_count'] / driver_combined['row_count'] * 100)
driver_combined = driver_combined.sort_values(by='percentage_above_teammate', ascending=False).reset_index(drop=True)

driver_combined.head(20)
```

```python
# Calculating Percentage above teammate

drivers_with_min_races_46 = driver_row_counts_corrected_sorted[driver_row_counts_corrected_sorted['row_count'] > 45]

# Teammate Comparison DataFrame
df_filtered = df[['driver', 'constructor_name', 'race_name', 'final_position', 'year']].copy()
df_filtered['final_position'] = pd.to_numeric(df_filtered['final_position'], errors='coerce').fillna(24)
df_filtered['unique_race'] = df_filtered['race_name'].astype(str) + "_" + df_filtered['year'].astype(str)


def calculate_better_finishes(group):
    group = group.sort_values('final_position').reset_index(drop=True)
    group['better_than_teammate'] = group['final_position'] < group['final_position'].shift(-1)
    return group


df_comparison = df_filtered.groupby(['unique_race', 'constructor_name']).apply(calculate_better_finishes)
driver_better_counts = df_comparison.groupby('driver')['better_than_teammate'].sum().reset_index()
driver_better_counts.columns = ['driver', 'better_than_teammate_count']

driver_combined = drivers_with_min_races_46.merge(driver_better_counts, on='driver', how='inner')

driver_combined['percentage_above_teammate'] = (driver_combined['better_than_teammate_count'] / driver_combined['row_count'] * 100)
driver_combined = driver_combined.sort_values(by='percentage_above_teammate', ascending=False).reset_index(drop=True)

driver_combined.head(10)
```

```python
# Creating Corrected points column

points_mapping = {
    1: 25,
    2: 18,
    3: 15,
    4: 12,
    5: 10,
    6: 8,
    7: 6,
    8: 4,
    9: 2,
    10: 1
}

df['corrected_points'] = df['final_position'].map(points_mapping).fillna(0).astype(int)
print(df[['driver', 'final_position', 'corrected_points']].head())
```

```python
# DF with corrected points

df.head(3)
```

```python
# Total corrected points and Total races

driver_performance_ratio = df.groupby('driver').agg(
    total_corrected_points = ('corrected_points', 'sum'),
    total_races = ('driver', 'count') 
).reset_index()

driver_performance_ratio = driver_performance_ratio[driver_performance_ratio['total_races'] > 45]

# Calculate the ratio of corrected points to total races
driver_performance_ratio['ratio_points_per_race'] = (
    driver_performance_ratio['total_corrected_points'] / (driver_performance_ratio['total_races'] * 25)
)

driver_performance_ratio = driver_performance_ratio.sort_values(by='ratio_points_per_race', ascending=False).reset_index(drop=True)

driver_performance_ratio.head(30)
```

```python
# Calculating the win ratio

df_filtered = df[['driver', 'constructor_name', 'race_name', 'final_position', 'year']].copy()
df_filtered['final_position'] = pd.to_numeric(df_filtered['final_position'], errors='coerce').fillna(24)
df_filtered['unique_race'] = df_filtered['race_name'].astype(str) + "_" + df_filtered['year'].astype(str)

df_filtered['best_position'] = df_filtered.groupby(['unique_race', 'constructor_name'])['final_position'].transform('min')
df_filtered['better_than_teammates'] = df_filtered['final_position'] == df_filtered['best_position']

driver_wins_against_teammates = df_filtered[df_filtered['better_than_teammates']].groupby('driver').size().reset_index(name='wins_against_teammates')
driver_appearance_count = df['driver'].value_counts().reset_index()
driver_appearance_count.columns = ['driver', 'total_appearances']

drivers_with_min_races = driver_appearance_count[driver_appearance_count['total_appearances'] > 45]
driver_wins_against_teammates = driver_wins_against_teammates.merge(
    drivers_with_min_races, on='driver', how='inner'
)

driver_wins_against_teammates['win_ratio'] = (
    driver_wins_against_teammates['wins_against_teammates'] /
    driver_wins_against_teammates['total_appearances']
)

driver_wins_against_teammates = driver_wins_against_teammates.sort_values(by='win_ratio', ascending=False).reset_index(drop=True)

print(driver_wins_against_teammates.head(20))
```

```python
# Combined dataframe of two ratios

combined_df = driver_performance_ratio.merge(
    driver_wins_against_teammates[['driver', 'win_ratio']],
    on='driver',
    how='inner'
)

combined_df['combined_metric'] = combined_df['ratio_points_per_race'] * combined_df['win_ratio']

combined_df = combined_df.sort_values(by='combined_metric', ascending=False).reset_index(drop=True)
combined_df[['driver', 'ratio_points_per_race', 'win_ratio', 'combined_metric']].head(30)
```
