```python
import pandas as pd
import itertools
import matplotlib.pyplot as plt
```

```python
def factorial(n): #defining a class which runs the factorial which we can call when runing the formula
  fact = n
  if n == 0:
    fact = 1
  else:
    for i in range(1, n):
      fact = fact * i
  return fact
```

```python
n_values = [8,20,25,40,50]
p_values = [0.15,0.25,0.53,0.74,0.99]

# Create a plot
plt.figure(figsize=(10, 6))

# Iterate over each combination of n and p
for i in range(len(n_values)):
    n = n_values[i]  # Get the n (number of trials) at the current index
    p = p_values[i]  # Get the corresponding p (probability of success) at the current index

    results = []  # store the binomial distribution results

    for x in range(0,n+1):
      num_ = factorial(n)
      denominator = factorial(x)*factorial((n-x))
      combinations = num_/(denominator)
      P_x = combinations*(p**x)*((1-p)**(n-x))

      results.append([x, P_x])
    df = pd.DataFrame(results, columns=['x', 'P(x)'])
    plt.plot(df['x'], df['P(x)'], label=f'n={n}, p={p}')


plt.xlabel('Number of tries (x)') # Add labels and title
plt.ylabel('P(x)')
plt.title('Binomial Distribution for Different n and p Values')

plt.legend()

plt.show()
```

Pattern an explanation:
For small

p values (like 0.15 or 0.25), most of the time, there are fewer successes, so the distribution focuses on lower numbers of successes.

As

p increases (like 0.53, 0.74, or 0.99), the distribution moves toward higher numbers of successes, meaning it's more likely to have more successes.
--> Moving to the right

When

p is very large (like 0.99), and the number of trials

n is also large (like 50), most of the results are very close to the maximum possible number of successes.

```python
import math
n_events = [10,30,30,60,60]
lambda_ = [0.5,1,1.5,2,20]
plt.figure(figsize=(10, 6))


for i in range(len(lambda_)):
    l = lambda_[i]  # Get the lambda
    n = n_events[i]  # Get the n (number of trials) at the current index


    results = []  # store the binomial distribution results

    for x in range(0,n+1):
      num_ = (math.exp(-l))*(l**x)
      denominator = factorial(x)
      P_x = num_/denominator
      results.append([x, P_x])

    df = pd.concat([pd.DataFrame(results, columns=['x', 'P(x)'])])

    plt.plot(df['x'], df['P(x)'], label=f'lambda = {l}')

plt.xlabel('Observation (x)') # Add labels and title
plt.ylabel('P(x)')
plt.title('Poisson distribution with lambda = '+str(lambda_))
plt.legend()
plt.show()
```

```python
import scipy.stats as st # Import libraries, this one is for stats
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters
mean = 3700 #euro
sd = 5000 #euro it is a mistake on the assignment
dist = st.norm(loc=mean, scale=sd)
#define our x axis
x = np.linspace(st.norm(mean, sd).ppf(.0001), st.norm(mean, sd).ppf(.99999))
#define y axis
y = st.norm(mean, sd).pdf(x) # PDF means probability distribution function

ppfs = {} # Create an empty dictionary
for ppf in [.50, .75, .90, .95, .99]: #the percentiles
    p = dist.ppf(ppf)
    ppfs.update({ppf*100: p})
ppfs

fig, ax = plt.subplots(figsize=(10, 4))
ax.plot(x, y, color='k')
for i, ppf in enumerate(ppfs):
    ax.axvline(ppfs[ppf], color=f'C{i}', label=f'{ppf:.2f}th: {ppfs[ppf]:.1f}') # To add the legend by using the saved dictionary
ax.legend()
plt.show()

```
