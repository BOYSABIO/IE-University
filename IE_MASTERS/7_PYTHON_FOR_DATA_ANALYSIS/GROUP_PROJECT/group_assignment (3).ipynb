Hint: remember that a probability is calculated as the number of times an event happened divided by the total number of events

```python
#With the data available, who is the fastest driver in Formula 1? (0.5 point)
#Based on average fastest lap speed
average_race_speed = df.groupby('driver')['fastest_lap_speed'].mean()

sorted_average_race_speed = average_race_speed.sort_values(ascending=False)

sorted_average_race_speed.head()
```

```python
#Based on fastest lap time

import pandas as pd
import numpy as np

def time_to_seconds(time_value):
    if pd.isnull(time_value):
        return np.nan
    if isinstance(time_value, (int, float)):
        return time_value
    time_str = str(time_value)

    if ':' in time_str:
        minutes, seconds = time_str.split(':')
        total_seconds = int(minutes) * 60 + float(seconds)
        return total_seconds
    else:
        # Handle cases without a colon (just seconds)
        return float(time_str)


df['fastest_lap_seconds'] = df['fastest_lap_time'].apply(time_to_seconds)
sorted_df = df[df['fastest_lap_seconds'].notna()]

print('All drivers and thier fastest times grouped by circuits:')
print(sorted_df.groupby(['circuit_name', 'driver'])['fastest_lap_seconds'].min().sort_values(ascending = True))
print()

sorted_df = df[df['fastest_lap_seconds'].notna()]

driver_lap_times = sorted_df.groupby(['circuit_name', 'driver'])['fastest_lap_seconds'].min().reset_index()

driver_lap_times['min_time_per_circuit'] = driver_lap_times.groupby('circuit_name')['fastest_lap_seconds'].transform('min')

fastest_drivers = driver_lap_times[driver_lap_times['fastest_lap_seconds'] == driver_lap_times['min_time_per_circuit']]

fastest_drivers = fastest_drivers.drop(columns='min_time_per_circuit')

print('The fastest driver and their fastest time grouped by circuit:')
sorted_fastest_drivers = fastest_drivers[['circuit_name', 'driver', 'fastest_lap_seconds']].sort_values(by = 'fastest_lap_seconds', ascending = True)
print(sorted_fastest_drivers.head(3))
print()

print('The number of times a drver appears as the fastest across all circuits:')
most_appearing_drivers = sorted_fastest_drivers['driver'].value_counts()
print(most_appearing_drivers.head(3))
print()

print(f'{most_appearing_drivers.idxmax()} appears the most as the fastest driver across {most_appearing_drivers.max()} different circuits.')
```

```python
# Which is the driver with the most podiums without a win (position 2 or 3)? (0.5 point)
podium_finishes = df[df['final_position'].isin([2, 3])]

drivers_no_wins = df.groupby('driver')['final_position'].apply(lambda x: 1 not in x.values)

podium_no_wins = podium_finishes[podium_finishes['driver'].isin(drivers_no_wins[drivers_no_wins].index)]

podium_counts = podium_no_wins['driver'].value_counts()

print(podium_counts.head(3))

print(f'{podium_counts.idxmax()} has the most podiums ({podium_counts.max()}) without a win.')
```

```python
#Calculate the historical probability of each country of having a driver in the podium (0.5 point)
print('When looking at a countries entire history of races:')
print(df.groupby('driver_nationality')['final_position'].apply(lambda x: ((x == 1) | (x == 2) | (x == 3)).sum() / x.size).sort_values(ascending=False).head(3))
print()

print('When looking at a countries grouped by the year to see overtime.')
print(df.groupby(['driver_nationality', 'year'])['final_position'].apply(lambda x: ((x == 1) | (x == 2) | (x == 3)).sum() / x.size))

print('Second Method:')
podium_finishes = df[df['final_position'].isin([1, 2, 3])]

country_podium_counts = podium_finishes.groupby('driver_nationality')['final_position'].count()

total_race_counts = df.groupby('driver_nationality')['final_position'].count()

country_podium_probability = (country_podium_counts / total_race_counts) * 100 

sorted_podium_probabilities = country_podium_probability.sort_values(ascending=False)

sorted_podium_probabilities.head(3)
```

```python
# Calculate the historical probability of each country of having a driver win a race (0.5 point)
print('When looking at their a countries entire history of races:')
print(df.groupby('driver_nationality')['final_position'].apply(lambda x: (x == 1).sum() / x.size).sort_values(ascending=False).head(3))
print()

print('When looking at a countries grouped by the year to see overtime.')
print(df.groupby(['driver_nationality', 'year'])['final_position'].apply(lambda x: (x == 1).sum() / x.size))
print()

print('Second Method:')
race_wins = df[df['final_position'] == 1]

country_race_win_counts = race_wins.groupby('driver_nationality')['final_position'].count()
total_race_counts = df.groupby('driver_nationality')['final_position'].count()

country_win_probability = (country_race_win_counts / total_race_counts) * 100 

sorted_win_probabilities = country_win_probability.sort_values(ascending=False)
print(sorted_win_probabilities.head(3))
```

```python
# Which driver was the youngest to win a race? (0.5 point)
df_wins = df[df['final_position'] == 1]

df_wins[['driver', 'final_position', 'age', 'year']].loc[df_wins['age'].idxmin()]
```

```python
# Which drivers are the current ones with the longest history in Formula 1? (0.5 point)
driver_career_dates = df.groupby('driver')['race_date'].agg(['min', 'max']).reset_index()
driver_career_dates.rename(columns = {'min': 'first_race', 'max': 'last_race'}, inplace = True)

driver_career_dates['career_duration_years'] = (driver_career_dates['last_race'] - driver_career_dates['first_race']).dt.days / 365.25

driver_career_dates['race_count'] = df.groupby('driver').size().reset_index(name = 'race_count')['race_count']

driver_career_dates['avg_race_per_year'] = driver_career_dates['race_count'] / driver_career_dates['career_duration_years'].replace(0, 1)

driver_career_dates.sort_values(by = 'avg_race_per_year', ascending = False)

# Filter Drivers with career lasting under a year

print('When basing it off of the range of the drivers first race and their last race:')
print(driver_career_dates[['driver', 'career_duration_years']].sort_values(by = 'career_duration_years', ascending = False).iloc[0])
print()

print('When basing it off of the number of races in their career:')
print(driver_career_dates[['driver', 'race_count']].sort_values(by = 'race_count', ascending = False).iloc[0])
print()

print('When basing it off of both having both the longest career and most races:')
print(driver_career_dates[['driver', 'career_duration_years', 'race_count']].sort_values(['career_duration_years', 'race_count'], ascending = False).iloc[0])
```

```python
# Which would you say is the toughest circuit in Formula 1? (0.5 point)
df_not_finished = df[df['final_position'].isna()]

dnf = df_not_finished.groupby('circuit_name')['status'].size().reset_index().sort_values(by = 'status', ascending = False)
print(f'The circuit with the highest number of DNFs: \n{dnf.iloc[0]}')
print()

dnf_data = df[df['final_position'].isna()]

circuit_dnf_counts = dnf_data.groupby('circuit_name')['final_position'].size()
circuit_race_counts = df.groupby('circuit_name')['final_position'].size()

circuit_dnf_rate = (circuit_dnf_counts / circuit_race_counts) * 100 
sorted_dnf_rate = circuit_dnf_rate.sort_values(ascending=False).reset_index().rename(columns = {'final_position': 'dnf_percentage'})
print(f'If we put that into percentges of DNFs compared to races: \n{sorted_dnf_rate.iloc[0]}')
```

```python
#Which circuit requires the most f1 experience to win? (0.5 point)
winning_drivers = df[df['final_position'] == 1]

driver_first_year = df.groupby('driver')['year'].min()
winning_drivers = winning_drivers.merge(driver_first_year, on='driver', how='left', suffixes=('', '_first'))
winning_drivers['career_span_at_win'] = winning_drivers['year'] - winning_drivers['year_first'] 

circuit_experience_at_win = winning_drivers.groupby('circuit_name')['career_span_at_win'].mean()
sorted_circuit_experience_at_win = circuit_experience_at_win.sort_values(ascending=False).reset_index().rename(columns = {'career_span_at_win': 'avg_career_span_at_win_years'})
print(f'The circuit requiring the most experience to place first:\n{sorted_circuit_experience_at_win.iloc[0]}')
```
