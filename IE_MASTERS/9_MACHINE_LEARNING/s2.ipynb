```python
import pandas as pd

# Define the file path
file_path = 'Holtravel.dat'

# Read the file into a pandas DataFrame
df_2 = pd.read_csv(file_path, delimiter=';')

# Display the first few rows of the DataFrame
df_2
```

```python
# Merge df_1 and df_2 on the 'HOLCODE' and 'holidaycode' columns
df_combined = pd.merge(df_1, df_2, left_on='HOLCODE', right_on='holidaycode')

# Display the first few rows of the combined DataFrame
df_combined.head()
```

```python
# Change the column name 'DOB' to 'Date of Birth'
df_combined.rename(columns={'DOB': 'Date of Birth'}, inplace=True)
df_combined
```

```python
# Turn TRAVDATE into a datetime object
df_combined['TRAVDATE'] = pd.to_datetime(df_combined['TRAVDATE'], dayfirst=True) # dayfirst=True because the date is in the format DD/MM/YYYY
df_combined
```

```python
import numpy as np
# df_combined['GENDER'] = df_combined['GENDER'].astype(int)
df_combined[df_combined['GENDER'] == ''] = np.nan
df_combined['GENDER'].value_counts()
```

```python
df_combined['GENDER'] = df_combined['GENDER'].replace({'1': 'Male', '0': 'Female'}).astype('category')
df_combined
```

```python
df_combined['ACCOMMODATION'].replace(np.nan, 'NA', inplace=True)
df_combined['ACCOMMODATION'].value_counts(dropna=False)
```

```python
df_combined['HOLCOST'].replace('$null$', np.nan, inplace=True)
df_combined['HOLCOST'].value_counts(dropna=False)
df_combined['HOLCOST'] = df_combined['HOLCOST'].astype(float)

Q1 = df_combined['HOLCOST'].quantile(0.25)
Q3 = df_combined['HOLCOST'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

print(lower, upper)

filtered_df = df_combined[(df_combined['HOLCOST'] >= lower) & (df_combined['HOLCOST'] <= upper)]

mean = filtered_df['HOLCOST'].mean()
std = filtered_df['HOLCOST'].std()

filtered_df['Z_HOLCOST'] = ((filtered_df['HOLCOST'] - mean) / std).round(3)
print(filtered_df[['HOLCOST', 'Z_HOLCOST']].head())
```

```python
df_combined
```

```python
df_combined['Date of Birth'] = pd.to_datetime(df_combined['Date of Birth'], dayfirst=True)

df_combined['month'] = df_combined['Date of Birth'].dt.month
df_combined
```

```python
from datetime import datetime

df_combined['age'] = datetime.now().year - df_combined['Date of Birth'].dt.year
df_combined
```

```python
df_age_over_18 = df_combined[df_combined['age'] >= 18]
df_age_over_18
```

```python
df_combined['COUNTRY'].value_counts()
df_combined['country_number'] = df_combined['COUNTRY'].replace({'SPAIN': 1, 'ITALY': 2, 'GERMANY': 3, 'GREECE': 4})
df_combined
```

```python
df_combined['WEEKEND_SPAIN'] = ((df_combined['TRAVDATE'].dt.dayofweek == 4) | (df_combined['TRAVDATE'].dt.dayofweek == 5)) & (df_combined['COUNTRY'] == 'SPAIN')
df_combined
```

```python
# Create the three typical transformations of HOLCOST variable
df_combined['HOLCOST_LN'] = np.log(df_combined['HOLCOST'])
df_combined['HOLCOST_SQRT'] = np.sqrt(df_combined['HOLCOST'])
df_combined['HOLCOST_INV'] = 1 / df_combined['HOLCOST']

# Display the first few rows to verify the transformations
df_combined[['HOLCOST', 'HOLCOST_LN', 'HOLCOST_SQRT', 'HOLCOST_INV']].head()
```

```python
import re

def extract_surname(name):
	# Use regex to find the last word which is not a title
	match = re.search(r'(\b\w+\b)$', name)
	return match.group(1) if match else ''

df_combined['Surname'] = df_combined['NAME'].apply(extract_surname)
df_combined[['NAME', 'Surname']].head()
```

```python
import numpy as np

# Create a RAND variable with random values
df_combined['RAND'] = np.random.rand(len(df_combined))

# Sort the DataFrame by the RAND variable
df_combined.sort_values(by='RAND', inplace=True)

# Mark cases with a RAND value greater than 0.2
df_combined['MARK'] = df_combined['RAND'] > 0.2

df_combined
```

```python
# Export the df_combined DataFrame to an Excel file
df_combined.to_excel('result_dataset.xlsx', index=False)
```

```python
# Fill np.nan values in HOLCOST with a specific value (e.g., 0) or drop the rows containing np.nan values
df_combined['HOLCOST'].fillna(0, inplace=True)

# Create a new discrete version of HOLCOST variable with fixed size
df_combined['HOLCOST_DISCRETE'] = pd.cut(df_combined['HOLCOST'], bins=10)

# Count the number of records that are equal or greater than 2000
count_ge_2000 = df_combined[df_combined['HOLCOST'] >= 2000].shape[0]

print(f"Number of records with HOLCOST >= 2000: {count_ge_2000}")
```

```python
# Combine GENDER and COUNTRY into a new column
df_combined['GENDER_COUNTRY'] = df_combined['GENDER'].astype(str) + '_' + df_combined['COUNTRY']

# Group by the new column and calculate the mean HOLCOST for each group
gender_country_group = df_combined.groupby('GENDER_COUNTRY')['HOLCOST'].mean()

# Display the results
print(gender_country_group)
```

```python
mean_holcost_gender_country = df_combined.groupby(['GENDER', 'COUNTRY'])['HOLCOST'].mean()
print(mean_holcost_gender_country)
```
