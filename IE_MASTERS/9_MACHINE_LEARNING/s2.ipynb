```python
import pandas as pd

# Define the file path
file_path = 'Holtravel.dat'

# Read the file into a pandas DataFrame
df_2 = pd.read_csv(file_path, delimiter=';')

# Display the first few rows of the DataFrame
df_2
```

```python
# Merge df_1 and df_2 on the 'HOLCODE' and 'holidaycode' columns
df_combined = pd.merge(df_1, df_2, left_on='HOLCODE', right_on='holidaycode')

# Display the first few rows of the combined DataFrame
df_combined.head()
```

```python
# Change the column name 'DOB' to 'Date of Birth'
df_combined.rename(columns={'DOB': 'Date of Birth'}, inplace=True)
df_combined
```

```python
# Turn TRAVDATE into a datetime object
df_combined['TRAVDATE'] = pd.to_datetime(df_combined['TRAVDATE'], dayfirst=True) # dayfirst=True because the date is in the format DD/MM/YYYY
df_combined
```

```python
import numpy as np

df_combined['GENDER'] = np.where(df_combined['GENDER'] == '1', 'M', 'F')
df_combined
```
