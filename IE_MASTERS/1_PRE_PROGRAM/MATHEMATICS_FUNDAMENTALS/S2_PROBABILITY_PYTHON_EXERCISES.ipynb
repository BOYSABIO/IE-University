```python
def simulate_dice_rolls(num_rolls):
    rolls = [random.randint(1, 6) for _ in range(num_rolls)]
    outcome_counts = [rolls.count(outcome) for outcome in range(1, 7)]
    probabilities = [count / num_rolls for count in outcome_counts]
    return probabilities

# Vary the number of rolls to see how the probabilities converge
num_rolls_list = [10, 100,1000, 100000]  # List of different numbers of rolls

for num_rolls in num_rolls_list:
    probabilities = simulate_dice_rolls(num_rolls)

    # Plot the probabilities
    plt.plot(range(1, 7), probabilities, label=f'{num_rolls} Rolls')

# Add labels and legend
plt.xlabel('Dice Outcome')
plt.ylabel('Probability')
plt.title('Convergence of Dice Roll Probabilities')
plt.legend()

# Show the plot
plt.show()
```

```python
#Change the number of flips

def simulate_coin_flips(num_flips):
    flips = [random.choice(['Heads', 'Tails']) for _ in range(num_flips)]
    probability_heads = flips.count('Heads') / num_flips
    probability_tails = flips.count('Tails') / num_flips
    return probability_heads, probability_tails

num_flips = 10000000
heads_prob, tails_prob = simulate_coin_flips(num_flips)
print(f"Probability of heads: {heads_prob}")
print(f"Probability of tails: {tails_prob}")
```

```python
def simulate_coin_flips(num_flips):
    flips = [random.choice(['Heads', 'Tails']) for _ in range(num_flips)]
    probability_heads = flips.count('Heads') / num_flips
    probability_tails = flips.count('Tails') / num_flips
    return probability_heads, probability_tails

# Vary the number of flips to see how the probabilities converge
num_flips_list = [10]  # List of different numbers of flips

for num_flips in num_flips_list:
    heads_prob, tails_prob = simulate_coin_flips(num_flips)

    # Plot the probabilities
    plt.plot(['Heads', 'Tails'], [heads_prob, tails_prob], marker='o', label=f'{num_flips} Flips')

# Add labels and legend
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.title('Convergence of Coin Flip Probabilities')
plt.legend()

# Show the plot
plt.show()
```
